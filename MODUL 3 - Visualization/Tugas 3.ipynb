```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
guided_data = pd.read_csv('train_features (guided).csv')
tugas_data = pd.read_csv('test_features (tugas).csv')

guided_info = guided_data.info()
tugas_info = tugas_data.info()

guided_preview = guided_data.head()
tugas_preview = tugas_data.head()

(guided_info, guided_preview, tugas_info, tugas_preview)
```

```python
guided_path = 'train_features (guided).csv'
tugas_path = 'test_features (tugas).csv'

def compare_datasets(guided_path, tugas_path):
    guided_data = pd.read_csv(guided_path)
    tugas_data = pd.read_csv(tugas_path)

    print("Guided Data Info:")
    print(guided_data.info())
    print("\nTugas Data Info:")
    print(tugas_data.info())

    print("\nGuided Data Preview:")
    print(guided_data.head())
    print("\nTugas Data Preview:")
    print(tugas_data.head())

    guided_columns = set(guided_data.columns)
    tugas_columns = set(tugas_data.columns)

    print("\nColumns difference:")
    print("In Guided only:", guided_columns - tugas_columns)
    print("In Tugas only:", tugas_columns - guided_columns)
    print("Common:", guided_columns.intersection(tugas_columns))

compare_datasets(guided_path, tugas_path)
```

**Identifikasi beberapa perbedaan karakteristik antara dataset "Guided" dan "Tugas".**

1. Jumlah baris:
   - Guided Data: 3817 entri
   - Tugas Data: 3818 entri
   Tugas Data memiliki satu baris lebih banyak dibandingkan Guided Data.

2. Jumlah kolom:
   - Guided Data: 16 kolom
   - Tugas Data: 17 kolom
   Tugas Data memiliki satu kolom tambahan.

3. Kolom tambahan:
   - Tugas Data memiliki kolom "ID" yang tidak ada di Guided Data.

4. Tipe data:
   - Kedua dataset memiliki tipe data yang sama untuk kolom-kolom yang sesuai.
   - Guided Data: 12 float64, 1 int64, 3 object
   - Tugas Data: 12 float64, 2 int64, 3 object
   Perbedaannya adalah Tugas Data memiliki satu kolom int64 tambahan (kolom ID).

5. Nilai null:
   - Terdapat perbedaan jumlah nilai non-null di beberapa kolom, yang menunjukkan perbedaan dalam kelengkapan data.

6. Kolom "tanggal_menjadi_anggota":
   - Guided Data: 1065 non-null
   - Tugas Data: 1059 non-null
   Guided Data memiliki sedikit lebih banyak data tanggal keanggotaan yang tidak null.

7. Range tahun kelahiran:
   - Guided Data: contoh yang ditampilkan berkisar antara 1950-1979
   - Tugas Data: contoh yang ditampilkan berkisar antara 1957-1974
   Namun, ini hanya berdasarkan sampel kecil dan mungkin tidak mewakili seluruh dataset.

8. Nilai-nilai ekstrem:
   - Terdapat perbedaan dalam nilai-nilai ekstrem untuk beberapa kolom. Misalnya, 'belanja_daging' di Tugas Data memiliki nilai yang jauh lebih tinggi dalam sampel yang ditampilkan.

9. Distribusi nilai:
   - Meskipun keduanya memiliki kolom yang sama (kecuali ID), distribusi nilai mungkin berbeda. Ini terlihat dari perbedaan dalam jumlah non-null values dan perbedaan dalam contoh nilai yang ditampilkan.

Kesimpulannya, kedua dataset memiliki struktur yang sangat mirip, terdapat perbedaan dalam jumlah baris, adanya kolom ID tambahan di Tugas Data, dan kemungkinan perbedaan dalam distribusi dan range nilai di berbagai kolom.

```python
# ANALISIS SATATISTIK

def analyze_data(guided_path, tugas_path):
    guided_data = pd.read_csv(guided_path)
    tugas_data = pd.read_csv(tugas_path)

    numeric_columns = guided_data.select_dtypes(include=['int64', 'float64']).columns

    for column in numeric_columns:
        if column in tugas_data.columns:
            plt.figure(figsize=(12, 6))

            # Histogram
            plt.subplot(2, 2, 1)
            sns.histplot(data=guided_data, x=column, kde=True, color="blue", label="Guided")
            sns.histplot(data=tugas_data, x=column, kde=True, color="red", label="Tugas")
            plt.title(f'Distribusi {column}')
            plt.legend()

            # Box plot
            plt.subplot(2, 2, 2)
            sns.boxplot(data=pd.melt(pd.DataFrame({
                'Guided': guided_data[column],
                'Tugas': tugas_data[column]
            })), x='variable', y='value')
            plt.title(f'Box Plot {column}')

            # Statistik deskriptif
            guided_stats = guided_data[column].describe()
            tugas_stats = tugas_data[column].describe()

            plt.subplot(2, 2, 3)
            plt.axis('off')
            plt.text(0, 1, f"Guided Stats:\n{guided_stats}", verticalalignment='top')

            plt.subplot(2, 2, 4)
            plt.axis('off')
            plt.text(0, 1, f"Tugas Stats:\n{tugas_stats}", verticalalignment='top')

            plt.tight_layout()
            plt.show()

    # Analisis korelasi
    plt.figure(figsize=(12, 10))
    sns.heatmap(guided_data[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Matriks Korelasi - Guided Data')
    plt.show()

    plt.figure(figsize=(12, 10))
    sns.heatmap(tugas_data[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Matriks Korelasi - Tugas Data')
    plt.show()

analyze_data(guided_path, tugas_path)
```

Kesimpulan dari analisis statistik bahwa dari kedua dataset tersebut memiliki nilai yang mirip di setiap variabelnya.
